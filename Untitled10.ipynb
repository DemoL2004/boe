```python
import re
import nltk
from  nltk.tokenize import word_tokenize
```

```python
at
at=at.lower()
at
```

```python
ct=re.sub('[^a-zA-z]',' ',at)
ct=re.sub('\s+',' ',ct)
ct
```

```python
sl=nltk.sent_tokenize(at)
sl
```

```python
tokens=word_tokenize(at)
```

```python
for i in tokens:
    if i in [".",":",",","(",")","%","[","]"]:
        tokens.remove(i)
tokens
```

```python
stopwords=nltk.corpus.stopwords.words('english')
wf={}
for word in tokens:
    if word not in stopwords:
        if word not in wf:
            wf[word]=1
            
        else:
            wf[word]+=1
wf
```

```python
import string
from nltk.probability import FreqDist
removethese=set(stopwords+list(string.punctuation)+list(string.digits))
```

```python
wfr=FreqDist(wf)
```

```python
wfr.plot(30,title='freqplot')
```

```python
from collections import Counter

```

```python
dictionary=Counter(wfr)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
cloud=WordCloud(max_font_size=80,colormap='hsv').generate_from_frequencies(dictionary)
cloud
```

```python
plt.figure(figsize=(16,12))
plt.imshow(cloud,interpolation='bilinear')
plt.axis("off")
plt.show()
```
