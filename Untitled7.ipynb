```python
def correlation(dataset,threshold):
    col_corr=set()
    cm=dataset.corr()
    for i in range(len(cm.columns)):
        for j in range(i):
            if abs(cm.iloc[i,j])>threshold:
                col_corr.add(cm.columns[i])
    return col_corr
```

```python
c=correlation(df,0.85)
len(c)
```

```python
#create box plot
fig,ax=plt.subplots(ncols=6,nrows=5,figsize=(20,20))
index=0
ax=ax.flatten()
for c in d.columns:
    sns.boxplot(y=col,data=d,ax=ax[index])
    index+=1
plt.tight_layout(pad=0.5,w_pad=0.7,h_pad=0.5)
```
