```python
d['rad'].fillna(d["rad"].mean(),inplace=True)
```

```python
d.isnull().sum()
```

```python
c=d.corr()
plt.figure(figsize=(15,10))
sns.heatmap(c,annot=True,cmap="coolwarm")
```

```python
def cm(da,thre):
    cs=set()
    cmm=da.corr()
    for i in range(len(cmm.columns)):
        for j in range(i):
            if abs(cmm.iloc[i,j]>thre):
                   cs.add(cmm.columns[i])
    return cs
```

```python
c=cm(d,0.7)
c
```
